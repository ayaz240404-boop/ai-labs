```python
import gzip, struct
import numpy as np

def load_images(path):
    with gzip.open(path, "rb") as f:
        _, n, r, c = struct.unpack(">IIII", f.read(16))
        return np.frombuffer(f.read(), dtype=np.uint8).reshape(n, r, c)

def load_labels(path):
    with gzip.open(path, "rb") as f:
        _, n = struct.unpack(">II", f.read(8))
        return np.frombuffer(f.read(n), dtype=np.uint8).copy()  # <- copy()

train_images = load_images("emnist/gzip/emnist-letters-train-images-idx3-ubyte.gz")
train_labels = load_labels("emnist/gzip/emnist-letters-train-labels-idx1-ubyte.gz")
test_images  = load_images("emnist/gzip/emnist-letters-test-images-idx3-ubyte.gz")
test_labels  = load_labels("emnist/gzip/emnist-letters-test-labels-idx1-ubyte.gz")

# ориентация + нормализация
train_images = np.flip(np.transpose(train_images, (0,2,1)), axis=2).astype("float32") / 255.0
test_images  = np.flip(np.transpose(test_images,  (0,2,1)), axis=2).astype("float32") / 255.0

# labels 1..26 -> 0..25 (не inplace тоже можно)
train_labels = train_labels - 1
test_labels  = test_labels - 1
```

```python
import matplotlib.pyplot as plt

letters = [chr(ord('A') + i) for i in range(26)]

print("__ТРЕНИРОВАЧНЫЕ ИЗОБРАЖЕНИЯ__\n")
fig = plt.figure(figsize=(10,10))
for i in range(1,11):
    ax = fig.add_subplot(5,5,i)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.imshow(train_images[i], cmap="gray")
    ax.set_title(letters[train_labels[i]])
plt.show()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical

train_labels_cat = to_categorical(train_labels, 26)
test_labels_cat  = to_categorical(test_labels,  26)

model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(26, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()
```

```python
model.fit(train_images, train_labels_cat, epochs=10)
```

```python
scores = model.evaluate(test_images, test_labels_cat, verbose=1)
print("Точность, %:", round(scores[1] * 100, 2))
```

```python
import numpy as np

pred = model.predict(test_images)

n = 5
plt.imshow(test_images[n], cmap="gray")
plt.show()

print("Предсказание:", letters[np.argmax(pred[n])])
print("Правильно   :", letters[test_labels[n]])
```
